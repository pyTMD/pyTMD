````python
import pyTMD
import timescale
import cartopy.crs as ccrs
import matplotlib.gridspec
import matplotlib.pyplot as plt

lon = -62.5
lat = -67.5
time = timescale.time.date_range("2025-01-01", "2025-01-31", 1, "h")

tide_h = pyTMD.compute.tide_elevations(
    lon,
    lat,
    time,
    model="GOT4.10_nc",
    crs="4326",
    standard="datetime",
    method="linear",
)
````

```{eval-rst}
- ``model``: a tide model available in the ``pyTMD`` model database
- ``crs``: coordinate reference system (``4326``: WGS84 Latitude / Longitude)
- ``standard``: time format (see :ref:`background on time standards <time-standards>`)
- ``method``: ``xarray`` interpolation method
```

````python
fig = plt.figure(figsize=(9, 3))
gs = matplotlib.gridspec.GridSpec(1, 2, height_ratios=[1], wspace=-0.15)

ax1 = plt.subplot(gs[0, 0])
ax1.plot(time, tide_h)
ax1.set_ylabel(f"Elevation [{tide_h.units}]")
ax1.set_title(f"Time Series at {lat:.1f}°N {lon:.1f}°E")
for label in ax1.get_xticklabels(which="major"):
    label.set(rotation=30, horizontalalignment="right")

projection = ccrs.Stereographic(
    central_longitude=0.0,
    central_latitude=-90,
    true_scale_latitude=-71.0,
)
ax2 = plt.subplot(gs[0, 1], projection=projection)
ax2.plot(lon, lat, "r*", transform=ccrs.PlateCarree())
ax2.set_xlim(-560.0 * 5e3, 560.0 * 5e3)
ax2.set_ylim(-560.0 * 5e3, 560.0 * 5e3)
ax2.coastlines("50m");
````
