```python
import pyTMD
import timescale
import matplotlib.pyplot as plt
lon = -62.5
lat = -67.5
time = timescale.time.date_range('2025-01-01', '2025-01-31', 1, 'h')
tide_h = pyTMD.compute.tide_elevations(lon, lat, time, model='GOT4.10_nc',
    crs='4326', type='time series', standard='datetime',  method='linear')
```

```python
plt.plot(time, tide_h)
plt.ylabel(f'Elevation [{tide_h.units}]')
plt.title(f'Time Series at {lat:.1f}°N {lon:.1f}°E')
for label in plt.gca().get_xticklabels(which='major'):
    label.set(rotation=30, horizontalalignment='right')
```
